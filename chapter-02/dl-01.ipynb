```python
import seaborn as sns
import numpy as np
import torch
```

```python
def f(x):
    return torch.pow((x - 2.0), 2)
```

```python
x_axis_vals = np.linspace(-10, 10, 100)
y_axis_vals_p = f(torch.tensor(x_axis_vals)).numpy()

sns.lineplot(x=x_axis_vals, y=y_axis_vals_p, label="$f(x)=(x-2)^2$")
```
